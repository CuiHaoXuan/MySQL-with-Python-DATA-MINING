```python
import pandas as pd
df = pd.read_csv('som_4cluster.csv',index_col =0)
def sample_class_show(y,fig_title='各类型占比',savepath='res.png'):
    '''
    绘制饼图,其中y是一维标签列表
    sample_class_show(y_1,savepath='before_resample_y.png')
    sample_class_show(y_resampled,savepath='resample_y.png')
    '''
    #%matplotlib inline
    import matplotlib.pyplot as plt
    from collections import Counter
    target_stats=Counter(y)
    labels=list(target_stats.keys())
    sizes=list(target_stats.values())
    explode=tuple([0.1] * len(target_stats))
    fig, ax=plt.subplots()
    ax.pie(sizes, explode=explode,labels=labels, shadow=True,autopct='%1.1f%%')
    ax.set_title(fig_title)
    ax.axis('equal')
    plt.savefig(savepath)
    
sample_class_show(df['cluster'].values,fig_title='201901-201906 Category ratio',savepath='res.png')
```
